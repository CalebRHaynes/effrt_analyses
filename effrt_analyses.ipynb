```python
import pandas as pd
import glob
import os
#import seaborn as sns
```

```python
#load all xlsx files into pandas dataframes 
monetary_dfs = [pd.read_excel(i) for i in glob.glob("data/monetary/*.xlsx")]
```

```python
for df in monetary_dfs:
    '''
    this iterates over all monetary dfs and extracts variables of interest.
    variables labelled 'series' are defining subsets of 
    '''
    completed_series = df['Completed'] == 1
    easy_series = df['Choice'] == 0
    hard_series = df['Choice'] == 1
    win_series = df['Procedure'] == 'Win'
    lose_series = df['Procedure'] == 'Lose'
    prob_12_series = df['Probability'] == 12
    prob_50_series = df['Probability'] == 50
    prob_88_series = df['Probability'] == 88
    hard_12_series = df.query('Probability == 12 & Choice == 1')
    hard_50_series = df.query('Probability == 50 & Choice == 1')
    hard_88_series = df.query('Probability == 88 & Choice == 1')
    easy_12_series = df.query('Probability == 12 & Choice == 0')
    easy_50_series = df.query('Probability == 50 & Choice == 0')
    easy_88_series = df.query('Probability == 88 & Choice == 0')
    RM_hard_12_series = hard_12_series['Amount']
    RM_hard_50_series = hard_50_series['Amount']
    RM_hard_88_series = hard_88_series['Amount']
    RM_easy_12_series = easy_12_series['Amount']
    RM_easy_50_series = easy_50_series['Amount']
    RM_easy_88_series = easy_88_series['Amount']




    trial_count = df['Trials'].max()
    completed_count = len(completed_series[completed_series == True].index)
    easy_count = len(easy_series[easy_series == True].index)
    hard_count = len(hard_series[hard_series == True].index)
    unchosen_count = trial_count - easy_count - hard_count
    win_count = len(win_series[win_series == True].index)
    lose_count = len(lose_series[lose_series == True].index)
    prob_12_count = len(prob_12_series[prob_12_series == True].index)
    prob_50_count = len(prob_50_series[prob_50_series == True].index)
    prob_88_count = len(prob_88_series[prob_88_series == True].index)
    hard_12_count = len(hard_12_series)
    hard_50_count = len(hard_50_series)
    hard_88_count = len(hard_88_series)
    easy_12_count = len(easy_12_series)
    easy_50_count = len(easy_50_series)
    easy_88_count = len(easy_88_series)


    completed_perc = completed_count / trial_count
    incomplete_perc = 1 - completed_perc  
    hard_prop = hard_count / trial_count
    easy_prop = easy_count / trial_count
    win_prop = win_count / trial_count
    lose_prop = lose_count / trial_count
    prob_12_prop = prob_12_count / trial_count
    prob_50_prop = prob_50_count / trial_count
    prob_88_prop = prob_88_count / trial_count
    hard_12_prop = hard_12_count / trial_count
    hard_50_prop = hard_50_count / trial_count
    hard_88_prop = hard_88_count / trial_count
    easy_12_prop = easy_12_count / trial_count
    easy_50_prop = easy_50_count / trial_count
    easy_88_prop = easy_88_count / trial_count


    overall_reward_amounts = df['Amount']
    overall_expectedValue_series = df['Amount'] * df['Probability'] / 100

    rt_series = df['Decision.RT']
    rt_avg = rt_series.sum() / trial_count

    tapSpeed_series = df['TapSpeed']
    tapSpeed_avg = tapSpeed_series.sum() / trial_count

    winnings_amt = df['Winnings'].sum()







```
